```python
ps.sqldf("""
SELECT df.Embarked,
df.Sex,
SUM(df.Age)AS total_age,
MAX(df.Age)AS max_age,
MIN(df.Age)AS min_age,
AVG(df.Age)AS avg_age,
COUNT(df.PassengerID)AS total_passengers,
COUNT(DISTINCT (df.Ticket))AS total_ticket
FROM titanic_disaster_dataset AS df
WHERE df.Age BETWEEN 18 AND 60
GROUP BY df.Embarked,df.Sex
HAVING avg_age >30
ORDER BY df.Embarked DESC , df.Sex
""")
```

```python
cities=pd.DataFrame({'code':['s','c','l'],'city':['southampton','cherbourg','london']})
```

```python
ps.sqldf("""
SELECT df.Embarked,
df.Sex,c.city,
SUM(df.Age)AS total_age,
MAX(df.Age)AS max_age,
MIN(df.Age)AS min_age,
AVG(df.Age)AS avg_age,
COUNT(df.PassengerID)AS total_passengers,
COUNT(DISTINCT (df.Ticket))AS total_ticket
FROM titanic_disaster_dataset AS df
INNER JOIN cities AS c ON df.Embarked=c.code
GROUP BY df.Embarked,df.Sex

""")
```

```python
ps.sqldf("""
SELECT df.Embarked,
df.Sex,c.city,
SUM(df.Age)AS total_age,
MAX(df.Age)AS max_age,
MIN(df.Age)AS min_age,
AVG(df.Age)AS avg_age,
COUNT(df.PassengerID)AS total_passengers,
COUNT(DISTINCT (df.Ticket))AS total_ticket
FROM titanic_disaster_dataset AS df
LEFT JOIN cities AS c ON df.Embarked=c.code
GROUP BY df.Embarked,df.Sex

""")
```

```python
ps.sqldf("""
SELECT df.Embarked,
df.Sex,c.city,
SUM(df.Age)AS total_age,
MAX(df.Age)AS max_age,
MIN(df.Age)AS min_age,
AVG(df.Age)AS avg_age,
COUNT(df.PassengerID)AS total_passengers,
COUNT(DISTINCT (df.Ticket))AS total_ticket
FROM cities AS c
LEFT JOIN titanic_disaster_dataset AS df ON df.Embarked=c.code
GROUP BY df.Embarked,df.Sex

""")
```
